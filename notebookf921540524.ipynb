<h2 style="color:#d63384; font-size:28px; font-weight:bold;">
Data loading and initial processing
</h2>

```python
# Load the dataset
import os
for dirname, _, filenames in os.walk('/kaggle/input/global-gdp-per-capita-1990-2023-world-bank'):
    for filename in filenames:
        if 'gdp' in filename.lower() or 'gdp-per-capita' in filename.lower():
            local_csv_path = os.path.join(dirname, filename)
            print(f"Found file: {local_csv_path}")
            break

try:
    df_raw = pd.read_csv(local_csv_path)
except:
    # Create sample data for testing
    print("Using sample data...")
    years = list(range(1990, 2024))
    countries = ['USA', 'China', 'India', 'Germany', 'Japan', 'UK', 'France', 'Brazil', 'Italy', 'Canada']
    data = []
    for country in countries:
        base_gdp = np.random.randint(10000, 50000)
        for year in years:
            growth = np.random.normal(0.02, 0.05)
            if year == 1990:
                gdp = base_gdp
            else:
                gdp = data[-1][3] * (1 + growth)
            data.append([country, country[:3], year, max(1000, gdp)])
    
    df_raw = pd.DataFrame(data, columns=['Entity', 'Code', 'Year', 'GDP per capita, PPP (constant 2021 international $)'])

df = df_raw.rename(columns={
    "Entity": "country",
    "Code": "iso3",
    "Year": "year",
    "GDP per capita, PPP (constant 2021 international $)": "gdp_per_capita"
})

df = df[["iso3", "country", "year", "gdp_per_capita"]]

df = df.dropna(subset=["gdp_per_capita"])

df["year"] = df["year"].astype(int)

df = df.sort_values(["country", "year"])

df["log_gdp"] = np.log1p(df["gdp_per_capita"])
df["gdp_growth"] = df.groupby("country")["gdp_per_capita"].pct_change() * 100

print(" Dataset Information:")
print(f"Shape: {df.shape}")
print(f"Time Range: {df['year'].min()} - {df['year'].max()}")
print(f"Number of Countries: {df['country'].nunique()}")
print("\nFirst 5 rows:")
display(df.head())
```

<h2 style="color:#d63384; font-size:28px; font-weight:bold;">
Exploratory Data Analysis (EDA)
</h2>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


print("Analyzing Global GDP Trends...")

global_gdp = df.groupby('year')['gdp_per_capita'].agg(['mean', 'median', 'std']).reset_index()

plt.figure(figsize=(12, 6))
plt.plot(global_gdp['year'], global_gdp['mean'], marker='o', linewidth=2)
plt.title("Global Average GDP Per Capita Over Time", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average GDP Per Capita", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


latest_year = df['year'].max()
top_countries = df[df['year'] == latest_year].nlargest(10, 'gdp_per_capita')

plt.figure(figsize=(12, 6))
plt.bar(top_countries['country'], top_countries['gdp_per_capita'])
plt.title(f"Top 10 Countries by GDP Per Capita ({latest_year})", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("GDP Per Capita ($)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


growth_data = df.dropna(subset=['gdp_growth'])
avg_growth = growth_data.groupby('country')['gdp_growth'].mean().reset_index()
avg_growth = avg_growth.sort_values('gdp_growth', ascending=False).head(15)

plt.figure(figsize=(12, 6))
plt.bar(avg_growth['country'], avg_growth['gdp_growth'])
plt.title("Top 15 Countries by Average Annual GDP Growth Rate", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Average Growth Rate (%)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))
plt.hist(df[df['year'] == latest_year]['gdp_per_capita'], bins=50, edgecolor='black')
plt.title(f"Distribution of GDP Per Capita ({latest_year})", fontsize=16)
plt.xlabel("GDP Per Capita ($)", fontsize=12)
plt.ylabel("Number of Countries", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Preparing Data for Modeling
</h2>

```python
print(" Preparing Data for Modeling...")

def prepare_country_data(country_df):
    """Prepare time series data for a single country"""
    country_df = country_df.sort_values('year').copy()
    
    country_df['year_norm'] = country_df['year'] - country_df['year'].min()
    
    for lag in [1, 2, 3]:
        country_df[f'gdp_lag_{lag}'] = country_df['gdp_per_capita'].shift(lag)
        if 'gdp_growth' in country_df.columns:
            country_df[f'growth_lag_{lag}'] = country_df['gdp_growth'].shift(lag)
    
    for window in [2, 3]:
        country_df[f'rolling_mean_{window}'] = country_df['gdp_per_capita'].rolling(window=window).mean()
        country_df[f'rolling_std_{window}'] = country_df['gdp_per_capita'].rolling(window=window).std()
    
    country_df['year_diff'] = country_df['gdp_per_capita'].diff()
    
    country_df['year_squared'] = country_df['year_norm'] ** 2
    
    return country_df

countries = df['country'].unique()[:30]  
country_data_dict = {}

for country in countries:
    country_df = df[df['country'] == country].copy()
    if len(country_df) >= 10: 
        prepared_df = prepare_country_data(country_df)
        prepared_df = prepared_df.dropna()
        if len(prepared_df) >= 5:  
            country_data_dict[country] = prepared_df

print(f" Prepared data for {len(country_data_dict)} countries")
```

<h2 style="color:#d63384; font-size:28px; font-weight:bold;">
Training Advanced ML Models</h2>

```python

class GDPForecaster:
    def __init__(self):
        self.models = {
            'RandomForest': RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42),
            'GradientBoosting': GradientBoostingRegressor(n_estimators=100, max_depth=5, random_state=42),
            'XGBoost': XGBRegressor(n_estimators=100, max_depth=5, random_state=42),
            'LightGBM': LGBMRegressor(n_estimators=100, max_depth=5, random_state=42),
            'Ridge': Ridge(alpha=1.0, random_state=42),
            'Lasso': Lasso(alpha=0.1, random_state=42)
        }
        self.results = {}
    
    def train_for_country(self, country_name, country_df):
        """Train models for a specific country"""
        if len(country_df) < 10:
            return None
        
        # Prepare features and target
        feature_cols = [col for col in country_df.columns if col not in 
                       ['country', 'iso3', 'year', 'gdp_per_capita', 'log_gdp', 'gdp_growth']]
        
        X = country_df[feature_cols].values
        y = country_df['gdp_per_capita'].values
        
        if len(X) < 5:
            return None
        
        tscv = TimeSeriesSplit(n_splits=min(3, len(X)-1))
        
        country_results = {}
        for model_name, model in self.models.items():
            try:
                # Cross-validation scores
                cv_scores = cross_val_score(model, X, y, cv=tscv, 
                                          scoring='neg_mean_absolute_error')
                mae_scores = -cv_scores
                
                # Train final model
                model.fit(X, y)
                
                # Make predictions
                y_pred = model.predict(X)
                
                # Calculate metrics
                mae = mean_absolute_error(y, y_pred)
                rmse = np.sqrt(mean_squared_error(y, y_pred))
                r2 = r2_score(y, y_pred)
                mape = np.mean(np.abs((y - y_pred) / np.maximum(y, 1))) * 100
                
                country_results[model_name] = {
                    'model': model,
                    'mae_cv': np.mean(mae_scores),
                    'mae': mae,
                    'rmse': rmse,
                    'r2': r2,
                    'mape': mape,
                    'feature_cols': feature_cols
                }
                
            except Exception as e:
                continue
        
        return country_results if country_results else None
    
    def forecast_future(self, country_name, country_df, years_ahead=10):
        """Forecast future GDP for a country"""
        if country_name not in self.results:
            return None
        
        best_model_info = max(self.results[country_name].items(), 
                            key=lambda x: x[1]['r2'])
        best_model_name, best_model_data = best_model_info
        
        model = best_model_data['model']
        feature_cols = best_model_data['feature_cols']
        
        last_year = country_df['year'].max()
        future_years = list(range(last_year + 1, last_year + years_ahead + 1))
        
        forecasts = []
        current_data = country_df.copy()
        
        for year in future_years:
            future_features = []
            
            for col in feature_cols:
                if 'lag' in col:
                    lag_num = int(col.split('_')[-1])
                    if lag_num <= len(current_data):
                        future_features.append(current_data['gdp_per_capita'].iloc[-lag_num])
                    else:
                        future_features.append(current_data['gdp_per_capita'].iloc[-1])
                elif 'mean' in col or 'std' in col:
                    window = int(col.split('_')[-1])
                    if len(current_data) >= window:
                        if 'mean' in col:
                            future_features.append(current_data['gdp_per_capita'].tail(window).mean())
                        else:
                            future_features.append(current_data['gdp_per_capita'].tail(window).std())
                    else:
                        future_features.append(current_data['gdp_per_capita'].iloc[-1])
                elif 'year_norm' in col:
                    min_year = current_data['year'].min()
                    future_features.append(year - min_year)
                elif 'year_squared' in col:
                    min_year = current_data['year'].min()
                    future_features.append((year - min_year) ** 2)
                elif 'year_diff' in col:
                    future_features.append(0)  
                else:
                    future_features.append(0)  
            
            future_gdp = model.predict([future_features])[0]
            
            forecasts.append({
                'country': country_name,
                'year': year,
                'gdp_per_capita': future_gdp,
                'model_used': best_model_name
            })
            
            
            new_gdp = future_gdp
        
        return pd.DataFrame(forecasts)

forecaster = GDPForecaster()

print("Training models...")
for country_name, country_df in country_data_dict.items():
    results = forecaster.train_for_country(country_name, country_df)
    if results:
        forecaster.results[country_name] = results

print(f" Trained models for {len(forecaster.results)} countries")
```

```python
# ---- Compute Growth Rate ----
growth_data = []

for country in forecast_df['country'].unique():
    country_data = forecast_df[forecast_df['country'] == country]
    if len(country_data) >= 2:
        start_gdp = country_data.iloc[0]['gdp_per_capita']
        end_gdp = country_data.iloc[-1]['gdp_per_capita']
        growth_rate = ((end_gdp - start_gdp) / start_gdp) * 100

        iso3_row = countries_info[countries_info['country'] == country]
        iso3 = iso3_row['iso3'].iloc[0] if not iso3_row.empty else None
        
        if iso3:
            growth_data.append({
                'country': country,
                'iso3': iso3,
                'growth_rate': growth_rate
            })

if growth_data:
    growth_df = pd.DataFrame(growth_data)

    # Merge with world map
    map2 = world.merge(growth_df[['iso3', 'growth_rate']], on="iso3", how="left")

    # Plot
    fig, ax = plt.subplots(figsize=(17, 9))
    map2.plot(
        column='growth_rate',
        cmap='RdYlGn',
        linewidth=0.4,
        edgecolor='black',
        legend=True,
        vmin=growth_df['growth_rate'].min() - 10,
        vmax=growth_df['growth_rate'].max() + 10,
        ax=ax
    )

    ax.set_title(f"Predicted GDP Growth Rate (2024‚Äì{final_year})", fontsize=20)
    ax.axis("off")
    plt.tight_layout()
    plt.show()

    print("Forecast growth map displayed.")
```
