```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta



tickers = {
    # High Risk / Distressed Companies
    "WE": {"name": "WeWork", "sector": "Real Estate", "status": "bankrupt"},
    "AMC": {"name": "AMC Entertainment", "sector": "Entertainment", "status": "distressed"},
    "BYND": {"name": "Beyond Meat", "sector": "Consumer Goods", "status": "distressed"},
    "PTON": {"name": "Peloton", "sector": "Fitness", "status": "distressed"},
    "FFIE": {"name": "Faraday Future", "sector": "Automotive", "status": "distressed"},
    
    # Stable Tech Companies
    "AAPL": {"name": "Apple", "sector": "Technology", "status": "stable"},
    "MSFT": {"name": "Microsoft", "sector": "Technology", "status": "stable"},
    "AMZN": {"name": "Amazon", "sector": "Technology", "status": "stable"},
    "GOOGL": {"name": "Google", "sector": "Technology", "status": "stable"},
    "NVDA": {"name": "Nvidia", "sector": "Technology", "status": "stable"},
    
    # Volatile Companies
    "TSLA": {"name": "Tesla", "sector": "Automotive", "status": "volatile"},
    "SNAP": {"name": "Snap Inc", "sector": "Technology", "status": "volatile"},
    "SPCE": {"name": "Virgin Galactic", "sector": "Aerospace", "status": "volatile"}
}

#  Enhanced Data Download Function

def download_enhanced_company_data(ticker, info):
    print(f"ðŸ“¥ Downloading {info['name']} ({ticker})...")
    
    company = yf.Ticker(ticker)
    company_info = company.info
    
    stock_data = company.history(period="5y", interval="1d")
    stock_data.reset_index(inplace=True)
    stock_data["Ticker"] = ticker
    stock_data["Company"] = info['name']
    stock_data["Sector"] = info['sector']
    stock_data["Status"] = info['status']
    
    # Calculate additional metrics
    if not stock_data.empty:
        stock_data['Daily_Return'] = stock_data['Close'].pct_change()
        stock_data['Volatility'] = stock_data['Daily_Return'].rolling(window=30).std()
        stock_data['MA_50'] = stock_data['Close'].rolling(window=50).mean()
        stock_data['MA_200'] = stock_data['Close'].rolling(window=200).mean()
    
    # Financial Data with error handling
    financial_data = {}
    
    try:
        # Income Statement
        income_stmt = company.financials.T
        income_stmt["Ticker"] = ticker
        income_stmt["Company"] = info['name']
        financial_data['income'] = income_stmt
    except:
        financial_data['income'] = pd.DataFrame()
    
    try:
        # Balance Sheet
        balance_sheet = company.balance_sheet.T
        balance_sheet["Ticker"] = ticker
        balance_sheet["Company"] = info['name']
        financial_data['balance'] = balance_sheet
    except:
        financial_data['balance'] = pd.DataFrame()
    
    try:
        # Cash Flow
        cash_flow = company.cashflow.T
        cash_flow["Ticker"] = ticker
        cash_flow["Company"] = info['name']
        financial_data['cashflow'] = cash_flow
    except:
        financial_data['cashflow'] = pd.DataFrame()
    
    # Company Info
    info_data = {
        'Ticker': ticker,
        'Company': info['name'],
        'Sector': info['sector'],
        'Status': info['status'],
        'Market Cap': company_info.get('marketCap', np.nan),
        'Employees': company_info.get('fullTimeEmployees', np.nan),
        'Country': company_info.get('country', 'N/A'),
        'Industry': company_info.get('industry', 'N/A')
    }
    
    return stock_data, financial_data, info_data



all_stock_data = []
all_income_statements = []
all_balance_sheets = []
all_cashflows = []
company_info_list = []

for ticker, info in tickers.items():
    try:
        stock, financials, info_data = download_enhanced_company_data(ticker, info)
        
        all_stock_data.append(stock)
        
        if not financials['income'].empty:
            all_income_statements.append(financials['income'])
        if not financials['balance'].empty:
            all_balance_sheets.append(financials['balance'])
        if not financials['cashflow'].empty:
            all_cashflows.append(financials['cashflow'])
            
        company_info_list.append(info_data)
        
    except Exception as e:
        print(f" Error downloading {ticker}: {str(e)}")



# Stock Data
if all_stock_data:
    df_stock_enhanced = pd.concat(all_stock_data, ignore_index=True)
    
    df_stock_enhanced['Price_Change'] = df_stock_enhanced['Close'] - df_stock_enhanced['Open']
    df_stock_enhanced['High_Low_Spread'] = df_stock_enhanced['High'] - df_stock_enhanced['Low']
    
    # Save enhanced stock data
    df_stock_enhanced.to_csv("enhanced_stock_data.csv", index=False)
    print(f" Saved enhanced stock data: {len(df_stock_enhanced)} rows")

def safe_concat(data_list, name):
    if data_list:
        result = pd.concat(data_list, ignore_index=True)
        result.to_csv(f"enhanced_{name}.csv", index=False)
        print(f" Saved {name}: {len(result)} rows")
        return result
    return pd.DataFrame()

df_income = safe_concat(all_income_statements, "income_statements")
df_balance = safe_concat(all_balance_sheets, "balance_sheets")
df_cashflow = safe_concat(all_cashflows, "cashflows")

df_company_info = pd.DataFrame(company_info_list)
df_company_info.to_csv("company_info.csv", index=False)

print(f" Company info saved: {len(df_company_info)} companies")


print("\n Creating Analysis Dataset...")

if not df_stock_enhanced.empty:
    latest_stock = df_stock_enhanced.sort_values(['Ticker', 'Date']).groupby('Ticker').last().reset_index()
    
    stock_performance = df_stock_enhanced.groupby('Ticker').agg({
        'Close': ['last', 'mean', 'std'],
        'Volume': 'mean',
        'Daily_Return': ['mean', 'std'],
        'Volatility': 'last'
    }).round(4)
    
    stock_performance.columns = ['Current_Price', 'Avg_Price', 'Price_Std', 
                                'Avg_Volume', 'Avg_Daily_Return', 'Return_Std', 'Current_Volatility']
    stock_performance = stock_performance.reset_index()
    
    analysis_df = pd.merge(df_company_info, stock_performance, on='Ticker', how='left')
    
    risk_mapping = {'stable': 0, 'volatile': 1, 'distressed': 2, 'bankrupt': 3}
    analysis_df['Risk_Score'] = analysis_df['Status'].map(risk_mapping)
    
    analysis_df.to_csv("company_analysis_dataset.csv", index=False)
    print(f" Analysis dataset created: {len(analysis_df)} companies")

print(" All enhanced data downloaded and processed successfully!")
```

```python

def calculate_financial_ratios(income_df, balance_df, cashflow_df):
    """Calculate advanced financial ratios for bankruptcy prediction"""
    
    ratios_list = []
    
    for ticker in df_analysis['Ticker']:
        latest_income = income_df[income_df['Ticker'] == ticker].iloc[-1] if not income_df[income_df['Ticker'] == ticker].empty else None
        latest_balance = balance_df[balance_df['Ticker'] == ticker].iloc[-1] if not balance_df[balance_df['Ticker'] == ticker].empty else None
        latest_cashflow = cashflow_df[cashflow_df['Ticker'] == ticker].iloc[-1] if not cashflow_df[cashflow_df['Ticker'] == ticker].empty else None
        
        if latest_income is not None and latest_balance is not None:
            try:
                gross_profit = latest_income.get('Gross Profit', np.nan)
                revenue = latest_income.get('Total Revenue', np.nan)
                net_income = latest_income.get('Net Income', np.nan)
                total_assets = latest_balance.get('Total Assets', np.nan)
                total_equity = latest_balance.get('Total Stockholder Equity', np.nan)
                current_assets = latest_balance.get('Current Assets', np.nan)
                current_liabilities = latest_balance.get('Current Liabilities', np.nan)
                total_liabilities = latest_balance.get('Total Liabilities', np.nan)
                
                gross_margin = gross_profit / revenue if revenue and revenue != 0 else np.nan
                net_margin = net_income / revenue if revenue and revenue != 0 else np.nan
                roa = net_income / total_assets if total_assets and total_assets != 0 else np.nan
                roe = net_income / total_equity if total_equity and total_equity != 0 else np.nan
                current_ratio = current_assets / current_liabilities if current_liabilities and current_liabilities != 0 else np.nan
                debt_to_equity = total_liabilities / total_equity if total_equity and total_equity != 0 else np.nan
                
                ratios_list.append({
                    'Ticker': ticker,
                    'Gross_Margin': gross_margin,
                    'Net_Margin': net_margin,
                    'ROA': roa,
                    'ROE': roe,
                    'Current_Ratio': current_ratio,
                    'Debt_to_Equity': debt_to_equity,
                    'Total_Assets': total_assets,
                    'Total_Revenue': revenue,
                    'Net_Income': net_income
                })
            except:
                continue
    
    return pd.DataFrame(ratios_list)

df_ratios = calculate_financial_ratios(df_income, df_balance, df_cashflow)
print(f" Calculated financial ratios for {len(df_ratios)} companies")
```

```python

print(" TRAINING BANKRUPTCY PREDICTION MODEL...")

feature_columns = [
    'Current_Price', 'Avg_Price', 'Price_Std', 'Avg_Volume', 
    'Avg_Daily_Return', 'Return_Std', 'Current_Volatility',
    'Gross_Margin', 'Net_Margin', 'ROA', 'ROE', 'Current_Ratio', 'Debt_to_Equity'
]

print(" Checking for NaN values in features...")

missing_summary = df_enhanced[feature_columns].isnull().sum()
print("Missing values per feature:")
for col, missing_count in missing_summary.items():
    missing_pct = (missing_count / len(df_enhanced)) * 100
    print(f"   {col}: {missing_count} missing ({missing_pct:.1f}%)")

columns_to_keep = []
for col in feature_columns:
    missing_pct = (df_enhanced[col].isnull().sum() / len(df_enhanced)) * 100
    if missing_pct < 80: 
        columns_to_keep.append(col)
    else:
        print(f" Removing {col} - {missing_pct:.1f}% missing")

print(f" Using {len(columns_to_keep)} features after cleaning")

X = df_enhanced[columns_to_keep].copy()

for col in columns_to_keep:
    col_median = X[col].median()
    if pd.isna(col_median):  
        col_median = 0
    X[col] = X[col].fillna(col_median)

y = df_enhanced['Bankruptcy_Risk']

print(f" Features after cleaning: {columns_to_keep}")
print(f" Target distribution: {y.value_counts().to_dict()}")
print(f" Final X shape: {X.shape}")
print(f" NaN values in X: {X.isnull().sum().sum()}")

if X.isnull().sum().sum() > 0:
    print(" WARNING: There are still NaN values in X!")
    print(X.isnull().sum())
else:
    print(" No NaN values remaining in X!")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f" Training set: {X_train.shape[0]} samples")
print(f" Test set: {X_test.shape[0]} samples")

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]

print(" Model trained successfully!")
print("\n MODEL PERFORMANCE:")
print(classification_report(y_test, y_pred, target_names=['Healthy', 'High Risk']))
```
