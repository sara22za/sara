<h1 style="color:#003366; font-weight:700;">Import Required Libraries
</h1>

```python
# Cell 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings('ignore')

# Modern Plot Setup & Style Configuration
from matplotlib.gridspec import GridSpec
import warnings
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

MODERN_COLORS = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD', '#FFA07A', '#20B2AA']
ATTRITION_COLORS = {'Yes': '#FF6B6B', 'No': '#4ECDC4'}
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD']

print(" Libraries imported successfully!")
```

<h1 style="color:#003366; font-weight:700;"> Load and Explore Dataset</h1>

```python
url = "https://raw.githubusercontent.com/IBM/employee-attrition-aif360/master/data/emp_attrition.csv"
df = pd.read_csv(url)

print(" DATASET LOADED SUCCESSFULLY!")
print(f"Dataset Shape: {df.shape}")
print("\n First 5 rows:")
print(df.head())

print("\n Dataset Info:")
print(df.info())

print("\n Target Variable Distribution:")
print(df['Attrition'].value_counts())
```

<h1 style="color:#003366; font-weight:700;">Data Preprocessing and Feature Engineering
</h1>

```python
data = df.copy()

data['Success'] = data['Attrition'].apply(lambda x: 0 if x == 'Yes' else 1)

columns_to_drop = ['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber']
data = data.drop(columns=columns_to_drop + ['Attrition'])

categorical_columns = data.select_dtypes(include=['object']).columns
label_encoders = {}

for col in categorical_columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))
    label_encoders[col] = le

print(" DATA PREPROCESSING COMPLETED!")
print(f"Remaining features: {data.shape[1]}")
print(f"Target variable distribution:\n{data['Success'].value_counts()}")
```

<h1 style="color:#003366; font-weight:700;">Advanced Feature Selection
</h1>

```python
X = data.drop('Success', axis=1)
y = data['Success']

rf_temp = RandomForestClassifier(n_estimators=100, random_state=42)
rf_temp.fit(X, y)

feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': rf_temp.feature_importances_
}).sort_values('importance', ascending=False)

print(" TOP 10 MOST IMPORTANT FEATURES:")
print(feature_importance.head(10))

selected_features = feature_importance.head(15)['feature'].values
X_selected = X[selected_features]

print(f"\n Selected {len(selected_features)} features for modeling")
```

<h1 style="color:#003366; font-weight:700;"> Attrition Distribution Plot
</h1>

```python
def create_attrition_distribution_plot(df):
    """
    Create modern distribution plots for employee attrition analysis
    """
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 12))
    fig.suptitle('ðŸš€ EMPLOYEE ATTRITION ANALYSIS - MODERN PLOTS', fontsize=20, fontweight='bold', y=0.95)
    
    # Plot 1: Attrition Distribution - Donut Chart
    attrition_counts = df['Attrition'].value_counts()
    wedges, texts, autotexts = ax1.pie(attrition_counts.values, 
                                      labels=attrition_counts.index,
                                      autopct='%1.1f%%',
                                      colors=[ATTRITION_COLORS['No'], ATTRITION_COLORS['Yes']],
                                      startangle=90,
                                      wedgeprops=dict(width=0.3))
    ax1.set_title('Employee Attrition Distribution\nDonut Chart', fontsize=14, fontweight='bold', pad=20)
    
    for autotext in autotexts:
        autotext.set_color('white')
        autotext.set_fontweight('bold')
        autotext.set_fontsize(12)
    
    if 'Age' in df.columns:
        sns.histplot(data=df, x='Age', hue='Attrition', multiple='stack', 
                    ax=ax2, palette=ATTRITION_COLORS, alpha=0.8)
        ax2.set_title('Age Distribution by Attrition Status', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Age')
        ax2.set_ylabel('Count')
    
    #  Monthly Income Distribution
    if 'MonthlyIncome' in df.columns:
        sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', ax=ax3, palette=ATTRITION_COLORS)
        ax3.set_title('Monthly Income Distribution by Attrition', fontsize=14, fontweight='bold')
        ax3.set_xlabel('Attrition Status')
        ax3.set_ylabel('Monthly Income ($)')
    
    #  Job Satisfaction by Attrition
    if 'JobSatisfaction' in df.columns:
        satisfaction_data = pd.crosstab(df['JobSatisfaction'], df['Attrition'])
        satisfaction_data.plot(kind='bar', ax=ax4, color=[ATTRITION_COLORS['No'], ATTRITION_COLORS['Yes']])
        ax4.set_title('Job Satisfaction Levels by Attrition', fontsize=14, fontweight='bold')
        ax4.set_xlabel('Job Satisfaction Level')
        ax4.set_ylabel('Count')
        ax4.legend(title='Attrition')
        plt.setp(ax4.xaxis.get_majorticklabels(), rotation=0)
    
    plt.tight_layout()
    plt.show()
    
    return fig

print(" CREATING MODERN ATTRITION DISTRIBUTION PLOTS...")
create_attrition_distribution_plot(df)
```

<h1 style="color:#003366; font-weight:700;">Correlation Heatmap Plot
</h1>

```python
def create_advanced_correlation_plot(data):
   
    numerical_cols = data.select_dtypes(include=[np.number]).columns
    corr_matrix = data[numerical_cols].corr()
    
    plt.figure(figsize=(16, 12))
    
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    
    sns.heatmap(corr_matrix, 
                mask=mask,
                annot=True, 
                cmap='RdBu_r', 
                center=0,
                square=True, 
                linewidths=0.5,
                cbar_kws={"shrink": 0.8},
                annot_kws={"size": 8})
    
    plt.title('ADVANCED CORRELATION MATRIX HEATMAP\nEmployee Dataset Features', 
              fontsize=18, fontweight='bold', pad=20)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

print(" CREATING ADVANCED CORRELATION HEATMAP PLOT...")
create_advanced_correlation_plot(df)
```

<h1 style="color:#003366; font-weight:700;">Feature Importance Plot
</h1>

```python
def create_feature_importance_plot(feature_importance, top_n=15):
   
    top_features = feature_importance.head(top_n)
    
    plt.figure(figsize=(14, 10))
    
    bars = plt.barh(top_features['feature'], top_features['importance'], 
                    color=MODERN_COLORS, alpha=0.8, edgecolor='black', linewidth=0.5)
    
    for i, (feature, importance) in enumerate(zip(top_features['feature'], top_features['importance'])):
        plt.text(importance + 0.001, i, f'{importance:.3f}', 
                va='center', fontsize=10, fontweight='bold')
    
    plt.xlabel('Feature Importance Score', fontsize=12, fontweight='bold')
    plt.ylabel('Features', fontsize=12, fontweight='bold')
    plt.title(' TOP FEATURE IMPORTANCE FOR EMPLOYEE SUCCESS PREDICTION', 
              fontsize=16, fontweight='bold', pad=20)
    plt.grid(axis='x', alpha=0.3)
    plt.gca().invert_yaxis()  
    
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    
    plt.tight_layout()
    plt.show()

print(" CREATING MODERN FEATURE IMPORTANCE PLOT...")
create_feature_importance_plot(feature_importance)
```

<h1 style="color:#003366; font-weight:700;"> Distribution Comparison Plots
</h1>

```python
def create_distribution_comparison_plots(df):
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 12))
    fig.suptitle(' ADVANCED DISTRIBUTION COMPARISON PLOTS', fontsize=20, fontweight='bold', y=0.95)
    
    #Age Distribution - Violin Plot
    if 'Age' in df.columns:
        sns.violinplot(data=df, x='Attrition', y='Age', ax=ax1, palette=ATTRITION_COLORS)
        ax1.set_title('Age Distribution by Attrition\nViolin Plot', fontsize=14, fontweight='bold')
        ax1.set_xlabel('Attrition Status')
        ax1.set_ylabel('Age')
    
    #  Monthly Income - Box Plot
    if 'MonthlyIncome' in df.columns:
        sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', ax=ax2, palette=ATTRITION_COLORS)
        ax2.set_title('Monthly Income Distribution\nBox Plot', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Attrition Status')
        ax2.set_ylabel('Monthly Income ($)')
    
    # Years at Company - KDE Plot
    if 'YearsAtCompany' in df.columns:
        for status in df['Attrition'].unique():
            subset = df[df['Attrition'] == status]
            sns.kdeplot(data=subset, x='YearsAtCompany', ax=ax3, 
                       label=status, fill=True, alpha=0.6, color=ATTRITION_COLORS[status])
        ax3.set_title('Years at Company Distribution\nKDE Plot', fontsize=14, fontweight='bold')
        ax3.set_xlabel('Years at Company')
        ax3.set_ylabel('Density')
        ax3.legend()
    
    # Job Satisfaction - Stacked Bar Plot
    if 'JobSatisfaction' in df.columns:
        satisfaction_cross = pd.crosstab(df['JobSatisfaction'], df['Attrition'])
        satisfaction_cross.plot(kind='bar', ax=ax4, color=[ATTRITION_COLORS['No'], ATTRITION_COLORS['Yes']])
        ax4.set_title('Job Satisfaction by Attrition\nStacked Bar Chart', fontsize=14, fontweight='bold')
        ax4.set_xlabel('Job Satisfaction Level')
        ax4.set_ylabel('Count')
        ax4.legend(title='Attrition')
        plt.setp(ax4.xaxis.get_majorticklabels(), rotation=0)
    
    plt.tight_layout()
    plt.show()

print(" CREATING DISTRIBUTION COMPARISON PLOTS...")
create_distribution_comparison_plots(df)
```

```python
def create_scatter_matrix_plot(df):
    
    key_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'TotalWorkingYears']
    available_features = [f for f in key_features if f in df.columns]
    
    if len(available_features) >= 2:
        pair_plot = sns.pairplot(df[available_features + ['Attrition']], 
                                hue='Attrition',
                                palette=ATTRITION_COLORS,
                                diag_kind='kde',
                                plot_kws={'alpha': 0.7, 's': 50},
                                diag_kws={'fill': True, 'alpha': 0.7})
        
        pair_plot.fig.suptitle(' MODERN SCATTER MATRIX - KEY EMPLOYEE FEATURES', 
                              fontsize=16, fontweight='bold', y=1.02)
        
        plt.show()
    else:
        print(" Not enough features available for scatter matrix")

print(" CREATING MODERN SCATTER MATRIX PLOT...")
create_scatter_matrix_plot(df)
```

<h1 style="color:#003366; font-weight:700;"> Categorical Analysis Plots
</h1>

```python
def create_categorical_analysis_plots(df):
    
    categorical_features = ['Department', 'EducationField', 'JobRole', 'Gender']
    available_cats = [f for f in categorical_features if f in df.columns]
    
    if len(available_cats) >= 2:
        fig, axes = plt.subplots(2, 2, figsize=(20, 12))
        axes = axes.ravel()
        fig.suptitle(' CATEGORICAL FEATURE ANALYSIS BY ATTRITION', fontsize=20, fontweight='bold', y=0.95)
        
        for i, feature in enumerate(available_cats[:4]):
            cross_tab = pd.crosstab(df[feature], df['Attrition'], normalize='index') * 100
            
            bars = cross_tab.plot(kind='bar', 
                                 ax=axes[i],
                                 color=[ATTRITION_COLORS['No'], ATTRITION_COLORS['Yes']],
                                 stacked=True,
                                 alpha=0.8)
            
            axes[i].set_title(f'{feature} - Attrition Rate', fontsize=14, fontweight='bold')
            axes[i].set_xlabel(feature)
            axes[i].set_ylabel('Percentage (%)')
            axes[i].legend(title='Attrition')
            axes[i].tick_params(axis='x', rotation=45)
            
            for container in axes[i].containers:
                axes[i].bar_label(container, fmt='%.1f%%', label_type='center', fontsize=9, fontweight='bold')
        
        for i in range(len(available_cats[:4]), 4):
            axes[i].set_visible(False)
        
        plt.tight_layout()
        plt.show()

print(" CREATING CATEGORICAL ANALYSIS PLOTS...")
create_categorical_analysis_plots(df)
```

<h1 style="color:#003366; font-weight:700;">Modern Statistical Summary Plots
</h1>

```python
def create_statistical_summary_plots(df):
   
    fig = plt.figure(figsize=(20, 12))
    gs = GridSpec(2, 2, figure=fig)
    
    fig.suptitle(' ADVANCED STATISTICAL SUMMARY PLOTS', fontsize=20, fontweight='bold', y=0.95)
    
    #  Numerical Features Distribution
    ax1 = fig.add_subplot(gs[0, 0])
    if 'MonthlyIncome' in df.columns:
        sns.histplot(data=df, x='MonthlyIncome', hue='Attrition', multiple='layer', 
                    ax=ax1, palette=ATTRITION_COLORS, alpha=0.7, kde=True)
        ax1.set_title('Monthly Income Distribution with KDE', fontsize=14, fontweight='bold')
        ax1.set_xlabel('Monthly Income ($)')
        ax1.set_ylabel('Density')
    
    #  Box plot for multiple numerical features
    ax2 = fig.add_subplot(gs[0, 1])
    numerical_features = ['Age', 'MonthlyIncome', 'YearsAtCompany']
    available_numerical = [f for f in numerical_features if f in df.columns]
    
    if available_numerical:
        normalized_data = df[available_numerical].apply(lambda x: (x - x.mean()) / x.std())
        normalized_data['Attrition'] = df['Attrition']
        
        melted_data = pd.melt(normalized_data, id_vars=['Attrition'], 
                             value_vars=available_numerical,
                             var_name='Feature', value_name='Normalized Value')
        
        sns.boxplot(data=melted_data, x='Feature', y='Normalized Value', hue='Attrition', 
                   ax=ax2, palette=ATTRITION_COLORS)
        ax2.set_title('Normalized Feature Distributions by Attrition', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Features')
        ax2.set_ylabel('Normalized Values')
        ax2.tick_params(axis='x', rotation=45)
    
    #  Correlation with target
    ax3 = fig.add_subplot(gs[1, :])
    if 'Success' in data.columns:
        numerical_data = data.select_dtypes(include=[np.number])
        target_corr = numerical_data.corr()['Success'].drop('Success').sort_values(ascending=True)
        
        bars = ax3.barh(target_corr.index, target_corr.values, 
                       color=MODERN_COLORS, alpha=0.8, edgecolor='black')
        
        for i, (feature, corr) in enumerate(zip(target_corr.index, target_corr.values)):
            ax3.text(corr + (0.01 if corr > 0 else -0.03), i, f'{corr:.3f}', 
                    va='center', fontsize=10, fontweight='bold',
                    color='white' if abs(corr) > 0.1 else 'black')
        
        ax3.set_title('Feature Correlation with Employee Success', fontsize=14, fontweight='bold')
        ax3.set_xlabel('Correlation Coefficient')
        ax3.axvline(x=0, color='black', linestyle='-', alpha=0.3)
        ax3.grid(axis='x', alpha=0.3)
    
    plt.tight_layout()
    plt.show()

print(" CREATING STATISTICAL SUMMARY PLOTS...")
create_statistical_summary_plots(df)
```

```python
def create_time_analysis_plots(df):
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 12))
    fig.suptitle(' TIME-BASED CAREER ANALYSIS PLOTS', fontsize=20, fontweight='bold', y=0.95)
    
    #  Years at Company vs Monthly Income
    if 'YearsAtCompany' in df.columns and 'MonthlyIncome' in df.columns:
        scatter = ax1.scatter(df['YearsAtCompany'], df['MonthlyIncome'], 
                             c=df['Attrition'].map(ATTRITION_COLORS), 
                             alpha=0.6, s=60, edgecolors='white', linewidth=0.5)
        ax1.set_title('Years at Company vs Monthly Income', fontsize=14, fontweight='bold')
        ax1.set_xlabel('Years at Company')
        ax1.set_ylabel('Monthly Income ($)')
        
        from matplotlib.lines import Line2D
        legend_elements = [Line2D([0], [0], marker='o', color='w', 
                                 markerfacecolor=ATTRITION_COLORS['No'], 
                                 markersize=10, label='Staying'),
                          Line2D([0], [0], marker='o', color='w', 
                                 markerfacecolor=ATTRITION_COLORS['Yes'], 
                                 markersize=10, label='Leaving')]
        ax1.legend(handles=legend_elements)
    
    #  Total Working Years Distribution
    if 'TotalWorkingYears' in df.columns:
        for status in df['Attrition'].unique():
            subset = df[df['Attrition'] == status]
            sns.kdeplot(data=subset, x='TotalWorkingYears', ax=ax2, 
                       label=status, fill=True, alpha=0.6, color=ATTRITION_COLORS[status])
        ax2.set_title('Total Working Years Distribution', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Total Working Years')
        ax2.set_ylabel('Density')
        ax2.legend()
    
    #  Years Since Last Promotion
    if 'YearsSinceLastPromotion' in df.columns:
        sns.boxplot(data=df, x='Attrition', y='YearsSinceLastPromotion', ax=ax3, palette=ATTRITION_COLORS)
        ax3.set_title('Years Since Last Promotion', fontsize=14, fontweight='bold')
        ax3.set_xlabel('Attrition Status')
        ax3.set_ylabel('Years Since Last Promotion')
    
    #  Years with Current Manager
    if 'YearsWithCurrManager' in df.columns:
        manager_data = pd.crosstab(df['YearsWithCurrManager'], df['Attrition'])
        manager_data.plot(kind='area', ax=ax4, color=[ATTRITION_COLORS['No'], ATTRITION_COLORS['Yes']], alpha=0.7)
        ax4.set_title('Years with Current Manager - Area Chart', fontsize=14, fontweight='bold')
        ax4.set_xlabel('Years with Current Manager')
        ax4.set_ylabel('Count')
        ax4.legend(title='Attrition')
    
    plt.tight_layout()
    plt.show()

print("â° CREATING TIME-BASED ANALYSIS PLOTS...")
create_time_analysis_plots(df)
```

<h1 style="color:#003366; font-weight:700;">Advanced Model Training- Setup
</h1>


```python
# Define models and parameters for GridSearch
models = {
    'Random Forest': {
        'model': RandomForestClassifier(random_state=42),
        'params': {
            'n_estimators': [100, 200],
            'max_depth': [10, 20, None],
            'min_samples_split': [2, 5]
        }
    },
    'Gradient Boosting': {
        'model': GradientBoostingClassifier(random_state=42),
        'params': {
            'n_estimators': [100, 200],
            'learning_rate': [0.05, 0.1],
            'max_depth': [3, 5]
        }
    },
    'XGBoost': {
        'model': XGBClassifier(random_state=42, eval_metric='logloss'),
        'params': {
            'n_estimators': [100, 200],
            'learning_rate': [0.05, 0.1],
            'max_depth': [3, 5]
        }
    },
    'SVM': {
        'model': SVC(random_state=42, probability=True),
        'params': {
            'C': [0.1, 1, 10],
            'kernel': ['rbf', 'linear']
        }
    }
}

print("MODEL TRAINING CONFIGURATION READY!")
print(f"Number of models: {len(models)}")
```

<h1 style="color:#003366; font-weight:700;"> Advanced Model Training - Execution
</h1>

```python
results = {}
best_models = {}

print(" TRAINING ADVANCED MACHINE LEARNING MODELS...\n")

for model_name, model_config in models.items():
    print(f" Training {model_name}...")
    
    grid_search = GridSearchCV(
        model_config['model'], 
        model_config['params'], 
        cv=5, 
        scoring='accuracy',
        n_jobs=-1,
        verbose=0
    )
    
    grid_search.fit(X_train_scaled, y_train)
    
    best_models[model_name] = grid_search.best_estimator_
    results[model_name] = {
        'best_score': grid_search.best_score_,
        'best_params': grid_search.best_params_
    }
    
    print(f"{model_name} completed! Best CV Score: {grid_search.best_score_:.4f}")
    print(f"   Best parameters: {grid_search.best_params_}\n")
```

<h1 style="color:#003366; font-weight:700;"> Model Evaluation and Comparison
</h1>

```python
evaluation_results = []

print(" MODEL EVALUATION ON TEST SET...\n")

for model_name, model in best_models.items():
    y_pred = model.predict(X_test_scaled)
    y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]
    
    accuracy = accuracy_score(y_test, y_pred)
    cv_scores = cross_val_score(model, X_train_scaled, y_train, cv=5)
    
    evaluation_results.append({
        'Model': model_name,
        'Test Accuracy': accuracy,
        'CV Mean Accuracy': cv_scores.mean(),
        'CV Std': cv_scores.std(),
        'Best Parameters': str(results[model_name]['best_params'])[:50] + "..."
    })
    
    print(f" {model_name}:")
    print(f"   Test Accuracy: {accuracy:.4f}")
    print(f"   CV Mean Accuracy: {cv_scores.mean():.4f} (Â±{cv_scores.std():.4f})")

results_df = pd.DataFrame(evaluation_results)
results_df = results_df.sort_values('Test Accuracy', ascending=False)

print("\n FINAL MODEL RANKINGS:")
print(results_df.to_string(index=False))
```
