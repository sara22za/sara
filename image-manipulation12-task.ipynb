# **Import Libraries**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
image = cv2.imread('/kaggle/input/birds222/WhatsApp Image 2025-04-13 at 23.01.16_28d18c49.jpg') 

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

# **filtering techniques**

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

blur = cv2.GaussianBlur(image, (9, 9), 0)

edges = cv2.Canny(image, 100, 200)

titles = ['Grayscale', 'Gaussian Blur', 'Edge Detection']
images = [gray, blur, edges]

plt.figure(figsize=(15,5))
for i in range(3):
    plt.subplot(1, 3, i+1)
    if i == 0:
        plt.imshow(images[i], cmap='gray')
    else:
        plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
cv2.imwrite('gray_image.jpg', gray)
cv2.imwrite('blur_image.jpg', blur)
cv2.imwrite('edges_image.jpg', edges)
```

# **image mainipulation**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image_path = '/kaggle/input/birds222/WhatsApp Image 2025-04-13 at 23.01.16_28d18c49.jpg'
image = cv2.imread(image_path)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

rotated = cv2.rotate(gray_image, cv2.ROTATE_90_CLOCKWISE)
cropped = gray_image[50:200, 50:200]
flipped = cv2.flip(gray_image, -1)

titles = ['Rotated', 'Cropped', 'Flipped']
images = [rotated, cropped, flipped]

plt.figure(figsize=(15,5))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

# **Image Filters**

```python
def preprocess_image(image):
  # Convert image to grayscale (if needed)
  gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  # Normalize image (if needed)
  normalized_image = gray_image.astype(np.float32) / 255.0

  # ... (add other preprocessing steps as needed)

  return normalized_image
```

```python
def mean_filtering(image, kernel_size):
  # Apply mean blur filter
  return cv2.blur(image, (kernel_size, kernel_size))
```

```python
def median_filtering(image, kernel_size):
  # Apply median blur filter
  return cv2.medianBlur(image, kernel_size)
```

```python
def gaussian_filtering(image, kernel_size, sigma):
  # Apply Gaussian blur filter
  return cv2.GaussianBlur(image, (kernel_size, kernel_size), sigma)
```

```python
# Apply different filtering techniques and store the results in separate variables
image=cv2.imread("/kaggle/input/checkboard/checkerboard_color.png")
preprocess_image=preprocess_image(image)
```

## *Modifying individual Channel*

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/kaggle/input/birds222/WhatsApp Image 2025-04-13 at 23.01.16_28d18c49.jpg')

if image is None:
    print("خطأ: لم يتم العثور على الصورة.")
else:
    img_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    h, s, v = cv2.split(img_hsv)

    h_new = cv2.add(h, 40)  

    img_NZ_merged = cv2.merge((h_new, s, v))

    img_NZ_rgb = cv2.cvtColor(img_NZ_merged, cv2.COLOR_HSV2RGB)

    plt.figure(figsize=[20, 5])
    plt.subplot(141); plt.imshow(h, cmap='gray'); plt.title("H Channel")
    plt.subplot(142); plt.imshow(s, cmap='gray'); plt.title("S Channel")
    plt.subplot(143); plt.imshow(v, cmap='gray'); plt.title("V Channel")
    plt.subplot(144); plt.imshow(img_NZ_rgb); plt.title("Modified Image")
    plt.show()
```
