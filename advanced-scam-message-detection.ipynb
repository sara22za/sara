    Text Preprocessing & Complexity Analysis Pipeline
  </h1>  
  <p style="color:#4aa398; font-size:18px; margin-top:5px;"></p>  
</div>

```python

# Download NLTK resources
print(" Downloading NLTK resources...")
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

class AdvancedTextPreprocessor:
    """Advanced text preprocessing with multiple techniques"""
    
    def __init__(self):
        self.stop_words = set(stopwords.words('english'))
        self.lemmatizer = WordNetLemmatizer()
        self.stemmer = PorterStemmer()
        
    def clean_text(self, text):
        """Comprehensive text cleaning"""
        if not isinstance(text, str):
            return ""
        
        text = text.lower()
        
        text = re.sub(r'[^a-zA-Z\s\.\?!,]', '', text)
        
        # Remove extra whitespace
        text = ' '.join(text.split())
        
        return text
    
    def tokenize_and_lemmatize(self, text):
        """Tokenize and lemmatize text"""
        tokens = word_tokenize(text)
        
        tokens = [token for token in tokens if token not in self.stop_words and len(token) > 2]
        
        tokens = [self.lemmatizer.lemmatize(token) for token in tokens]
        
        return ' '.join(tokens)
    
    def analyze_text_complexity(self, text):
        """Analyze text complexity with multiple metrics"""
        if not isinstance(text, str):
            return {
                'word_count': 0, 'sentence_count': 0, 'avg_sentence_length': 0,
                'lexical_diversity': 0, 'polarity': 0, 'subjectivity': 0
            }
            
        words = word_tokenize(text)
        sentences = text.split('.')
        
        # Basic metrics
        word_count = len(words)
        sentence_count = max(1, len([s for s in sentences if s.strip()]))
        avg_sentence_length = word_count / sentence_count
        
        unique_words = len(set(words))
        lexical_diversity = unique_words / max(1, word_count)
        
        sentiment = TextBlob(text).sentiment
        
        return {
            'word_count': word_count,
            'sentence_count': sentence_count,
            'avg_sentence_length': avg_sentence_length,
            'lexical_diversity': lexical_diversity,
            'polarity': sentiment.polarity,
            'subjectivity': sentiment.subjectivity
        }
    
    def advanced_preprocess(self, text):
        """Complete preprocessing pipeline"""
        cleaned_text = self.clean_text(text)
        processed_text = self.tokenize_and_lemmatize(cleaned_text)
        return processed_text

preprocessor = AdvancedTextPreprocessor()

print(" PROCESSING TEXT DATA...")

df_enhanced = df.copy()

print(" Analyzing text complexity...")
scam_analysis = df_enhanced['scammer_message'].apply(preprocessor.analyze_text_complexity).apply(pd.Series)
response_analysis = df_enhanced['user_response'].apply(preprocessor.analyze_text_complexity).apply(pd.Series)

df_enhanced = pd.concat([df_enhanced, scam_analysis.add_prefix('scam_'), 
                        response_analysis.add_prefix('response_')], axis=1)

print(" Preprocessing text data...")
df_enhanced['scam_processed'] = df_enhanced['scammer_message'].apply(preprocessor.advanced_preprocess)
df_enhanced['response_processed'] = df_enhanced['user_response'].apply(preprocessor.advanced_preprocess)

df_enhanced['combined_text'] = df_enhanced['scam_processed'] + ' ' + df_enhanced['response_processed']

print(" TEXT PROCESSING COMPLETED!")
print(f"Dataset shape: {df_enhanced.shape}")
print(f"Sample processed scam message: {df_enhanced['scam_processed'].iloc[0][:100]}...")
print(f"Sample processed response: {df_enhanced['response_processed'].iloc[0][:100]}...")

print("\n ENHANCED DATASET COLUMNS:")
print(df_enhanced.columns.tolist())
```

```python

from scipy.sparse import hstack
import joblib

print("LOADING MODELS FOR PREDICTION...")

try:
    logistic_model = joblib.load('scam_detection_logistic_regression.pkl')
    rf_model = joblib.load('scam_detection_random_forest.pkl')
    print(" Models loaded successfully!")
except:
    print(" Models not found. Using in-memory models...")
    logistic_model = selected_models['Logistic Regression']
    rf_model = selected_models['Random Forest']

class ScamPredictor:
    def __init__(self, logistic_model, rf_model, tfidf_scam, tfidf_response, safe_numerical_features, preprocessor):
        self.logistic_model = logistic_model
        self.rf_model = rf_model
        self.tfidf_scam = tfidf_scam
        self.tfidf_response = tfidf_response
        self.safe_numerical_features = safe_numerical_features
        self.preprocessor = preprocessor
    
    def predict_single(self, scam_message, user_response):
        """Predict success for a single scam message and user response"""
        
        scam_processed = self.preprocessor.advanced_preprocess(scam_message)
        response_processed = self.preprocessor.advanced_preprocess(user_response)
        
        scam_analysis = self.preprocessor.analyze_text_complexity(scam_message)
        response_analysis = self.preprocessor.analyze_text_complexity(user_response)
        
        scam_tfidf = self.tfidf_scam.transform([scam_processed])
        response_tfidf = self.tfidf_response.transform([response_processed])
        
        numerical_features = []
        for feature in self.safe_numerical_features:
            if feature in scam_analysis:
                numerical_features.append(scam_analysis[feature])
            elif feature in response_analysis:
                numerical_features.append(response_analysis[feature])
            else:
                numerical_features.append(0)  
        
        features_combined = hstack([scam_tfidf, response_tfidf, [numerical_features]])
        
        lr_pred = self.logistic_model.predict(features_combined)[0]
        lr_proba = self.logistic_model.predict_proba(features_combined)[0][1]
        
        rf_pred = self.rf_model.predict(features_combined)[0]
        rf_proba = self.rf_model.predict_proba(features_combined)[0][1]
        
        ensemble_proba = (lr_proba + rf_proba) / 2
        ensemble_pred = 1 if ensemble_proba > 0.5 else 0
        
        return {
            'logistic_regression': {
                'prediction': 'SUCCESS' if lr_pred == 1 else 'FAILURE',
                'confidence': lr_proba,
                'color': 'green' if lr_pred == 1 else 'red'
            },
            'random_forest': {
                'prediction': 'SUCCESS' if rf_pred == 1 else 'FAILURE',
                'confidence': rf_proba,
                'color': 'green' if rf_pred == 1 else 'red'
            },
            'ensemble': {
                'prediction': 'SUCCESS' if ensemble_pred == 1 else 'FAILURE',
                'confidence': ensemble_proba,
                'color': 'green' if ensemble_pred == 1 else 'red'
            },
            'scam_message': scam_message,
            'user_response': user_response
        }

predictor = ScamPredictor(logistic_model, rf_model, tfidf_scam_safe, tfidf_response_safe, safe_numerical_features, preprocessor)

print(" PREDICTOR READY!")
```
